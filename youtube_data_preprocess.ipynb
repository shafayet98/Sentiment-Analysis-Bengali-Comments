```python
import numpy as np
import pandas as pd
import re
```

```python
df = pd.read_csv("comment_final.csv")
```

```python
lst_cmt = df['text'].to_list()
```

```python
lst_cmt[1:100]
```

```python
# remove Html tag
cmt = []
for item in lst_cmt:
    try:
        cmt.append(re.sub('<.*?>','', item))
    except:
        print("Issue")
```

```python
def remove_punc(lines):
    punctuations = '''!()-[]ред{};:'"\,<>./?@#$%^&*_~'''
    punch_remove_line = []
    # remove punctuation from the string
    for i in range(len(lines)):
        no_punct = ""
        for char in lines[i]:
            if char not in punctuations:
                no_punct = no_punct + char
        punch_remove_line.append(no_punct)
    return punch_remove_line
```

```python
final_cmt_1 = remove_punc(cmt)
```

```python
# remove double spaces
final_cmt_2 = []
for item in final_cmt_1:
    try:
        final_cmt_2.append(re.sub('  ',' ', item))
    except:
        print("Isuue")
```

```python
final_cmt_2[:100]
```

```python
def create_dataset_csv(lines, filename):
    df = pd.DataFrame(lines, columns=["text"])
    df.to_csv(filename, index=False)
```

```python
create_dataset_csv(final_cmt_2,"youtube_comment.csv")
```

```python
def create_text_file(comments, filename):
    f = open(filename,"a")
    for cmt in comments:
        f.write(cmt + '\n')
    f.close()
```

```python
create_text_file(final_cmt_2,"youtube_comment.txt")
```
