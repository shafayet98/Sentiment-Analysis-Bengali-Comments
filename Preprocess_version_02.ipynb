```python
import numpy as np
import pandas as pd
import re
import string

import nltk
from bangla_stemmer.stemmer import stemmer
from nltk.tokenize import WhitespaceTokenizer
from nltk.tokenize import TweetTokenizer
from sklearn.model_selection import train_test_split

import matplotlib.pyplot as plt
```

```python
def read_files(filename):
    with open(filename, 'r', encoding='utf-8') as file:
        lines = []
        for line in file:
            lines.append(line)
        file.close()
    return lines
```

```python
negative_lines = read_files(filename = "all_negative.txt")
positive_lines = read_files(filename = "all_positive.txt")
```

```python
print(len(positive_lines))
print(len(negative_lines))
```

```python
positive_lines_ours = read_files(filename = "all_pos_ours.txt")
negative_lines_ours = read_files(filename = "all_neg_ours.txt")
```

```python
print(len(positive_lines_ours))
print(len(negative_lines_ours))
```

```python
data = data2 = ""
  
# Reading data from file1
with open('all_positive.txt', encoding='utf-8') as fp: 
    data = fp.read()
    
fp.close()

with open('all_pos_ours.txt', encoding='utf-8') as fp: 
    data2 = fp.read()
    
fp.close()
    
# To add the data of file2
# from next line
# data += "\n"
data += data2
  
with open ('final_pos_merged.txt', 'w', encoding='utf-8') as fp:
    fp.write(data)
```
